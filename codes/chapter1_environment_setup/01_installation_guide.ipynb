## 1. 安装前准备

### 1.1 检查 Python 版本

PyTorch 推荐使用 Python 3.8-3.11。

````python
import sys

print("=" * 50)
print("Python 环境检查")
print("=" * 50)
print(f"Python 版本: {sys.version}")
print(f"Python 路径: {sys.executable}")

# 检查版本是否合适
version_info = sys.version_info
if version_info.major == 3 and 8 <= version_info.minor <= 11:
    print("\n✅ Python 版本合适，可以安装 PyTorch")
else:
    print(f"\n⚠️  当前 Python 版本 {version_info.major}.{version_info.minor}，推荐使用 3.8-3.11")
````

### 1.2 选择安装方式

PyTorch 支持两种主要的安装方式：

| 方式 | 优点 | 缺点 | 推荐场景 |
|------|------|------|----------|
| **Conda** | 依赖管理好，环境隔离 | 下载较慢 | 推荐新手使用 |
| **Pip** | 安装快速，通用性好 | 可能有依赖冲突 | 熟悉 Python 的用户 |

### 1.3 选择 PyTorch 版本

访问 PyTorch 官网选择器：https://pytorch.org/get-started/locally/

需要选择：
- **操作系统**：Linux / Mac / Windows
- **包管理器**：Conda / Pip
- **Python 版本**：推荐 3.8-3.11
- **计算平台**：
  - **CPU**：没有 GPU 或只是学习
  - **CUDA 11.8**：NVIDIA GPU（较旧驱动）
  - **CUDA 12.1**：NVIDIA GPU（新驱动）
  - **ROCm**：AMD GPU

## 2. 安装命令

### ⚠️ 注意：以下命令在 Jupyter Notebook 中无法直接运行

请在**终端**中运行相应的安装命令。

### 2.1 使用 Conda 安装（推荐）

#### 创建虚拟环境
```bash
# 创建名为 pytorch_env 的环境，Python 3.11
conda create -n pytorch_env python=3.11

# 激活环境
conda activate pytorch_env
```

#### CPU 版本
```bash
conda install pytorch torchvision torchaudio cpuonly -c pytorch
```

#### GPU 版本（CUDA 11.8）
```bash
conda install pytorch torchvision torchaudio pytorch-cuda=11.8 -c pytorch -c nvidia
```

#### GPU 版本（CUDA 12.1）
```bash
conda install pytorch torchvision torchaudio pytorch-cuda=12.1 -c pytorch -c nvidia
```

### 2.2 使用 Pip 安装

#### CPU 版本
```bash
pip3 install torch torchvision torchaudio
```

#### GPU 版本（CUDA 11.8）
```bash
pip3 install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
```

#### GPU 版本（CUDA 12.1）
```bash
pip3 install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121
```

## 3. 验证安装

### 3.1 基本验证

````python
# 尝试导入 PyTorch
try:
    import torch
    print("✅ PyTorch 导入成功！")
    print(f"\nPyTorch 版本: {torch.__version__}")
except ImportError as e:
    print("❌ PyTorch 导入失败！")
    print(f"错误信息: {e}")
    print("\n请先安装 PyTorch，参考上面的安装命令。")
````

### 3.2 检查 CUDA 支持（如果安装了 GPU 版本）

````python
import torch

print("=" * 50)
print("CUDA 支持检查")
print("=" * 50)
print(f"CUDA 可用: {torch.cuda.is_available()}")

if torch.cuda.is_available():
    print(f"CUDA 版本: {torch.version.cuda}")
    print(f"GPU 数量: {torch.cuda.device_count()}")
    print(f"当前 GPU: {torch.cuda.current_device()}")
    print(f"GPU 名称: {torch.cuda.get_device_name(0)}")
    print("\n✅ CUDA 已正确配置！")
else:
    print("\n⚠️  CUDA 不可用（可能是 CPU 版本或驱动问题）")
````

### 3.3 检查 MPS 支持（Apple Silicon Mac）

````python
import torch

print("=" * 50)
print("MPS 支持检查 (Apple Silicon)")
print("=" * 50)

if hasattr(torch.backends, 'mps'):
    print(f"MPS 可用: {torch.backends.mps.is_available()}")
    print(f"MPS 已构建: {torch.backends.mps.is_built()}")
    
    if torch.backends.mps.is_available():
        print("\n✅ MPS 已正确配置！可以使用 Apple Silicon GPU 加速")
    else:
        print("\n⚠️  MPS 不可用（可能不是 Apple Silicon Mac）")
else:
    print("⚠️  当前 PyTorch 版本不支持 MPS")
    print("   需要 PyTorch 1.12+ 和 macOS 12.3+")
````

### 3.4 完整的设备检查

````python
import torch

print("=" * 60)
print("PyTorch 完整设备检查")
print("=" * 60)

print(f"\n📦 PyTorch 版本: {torch.__version__}")
print(f"🐍 Python 版本: {torch.__version__}")

print("\n💻 可用设备:")
print(f"  - CPU: ✅ (始终可用)")
print(f"  - CUDA: {'✅' if torch.cuda.is_available() else '❌'}")

if hasattr(torch.backends, 'mps'):
    print(f"  - MPS: {'✅' if torch.backends.mps.is_available() else '❌'}")

# 确定推荐使用的设备
print("\n🎯 推荐使用的设备:")
if torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"  {device} (NVIDIA GPU)")
    print(f"  GPU 名称: {torch.cuda.get_device_name(0)}")
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    device = torch.device("mps")
    print(f"  {device} (Apple Silicon GPU)")
else:
    device = torch.device("cpu")
    print(f"  {device} (CPU)")

print("\n" + "=" * 60)
print("✅ 安装验证完成！")
print("=" * 60)
````

### 3.5 简单功能测试

````python
import torch

print("=" * 50)
print("PyTorch 功能测试")
print("=" * 50)

# 测试 1：创建张量
print("\n测试 1: 创建张量")
x = torch.tensor([1, 2, 3, 4, 5])
print(f"张量 x: {x}")
print(f"形状: {x.shape}")
print(f"数据类型: {x.dtype}")
print("✅ 张量创建成功")

# 测试 2：基本运算
print("\n测试 2: 基本运算")
y = x * 2
print(f"x * 2 = {y}")
z = x + y
print(f"x + y = {z}")
print("✅ 基本运算正常")

# 测试 3：矩阵运算
print("\n测试 3: 矩阵运算")
a = torch.randn(3, 4)
b = torch.randn(4, 2)
c = torch.mm(a, b)
print(f"矩阵 A 形状: {a.shape}")
print(f"矩阵 B 形状: {b.shape}")
print(f"A @ B 形状: {c.shape}")
print("✅ 矩阵运算正常")

print("\n" + "=" * 50)
print("✅ 所有功能测试通过！")
print("=" * 50)
````

## 4. 常见问题

### Q1: 如何查看我的 CUDA 版本？

在终端运行：
```bash
nvidia-smi  # 查看驱动支持的最高 CUDA 版本
nvcc --version  # 查看已安装的 CUDA toolkit 版本（如果安装了）
```

### Q2: ImportError: cannot import name 'XXX'

**解决方案：** 卸载后重新安装，确保版本匹配
```bash
pip uninstall torch torchvision torchaudio
pip install torch torchvision torchaudio
```

### Q3: CUDA out of memory

这是运行时错误，不是安装问题。解决方法：
- 减小 batch size
- 使用梯度累积
- 使用混合精度训练

### Q4: MPS 不可用

**要求：**
- Apple Silicon Mac (M1/M2/M3)
- macOS 12.3 或更高版本
- PyTorch 1.12 或更高版本

### Q5: 安装很慢怎么办？

**使用国内镜像源：**

Pip 镜像：
```bash
pip install torch torchvision torchaudio -i https://pypi.tuna.tsinghua.edu.cn/simple
```

Conda 镜像：
```bash
# 添加清华源
conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/main/
conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/pkgs/free/
conda config --add channels https://mirrors.tuna.tsinghua.edu.cn/anaconda/cloud/pytorch/
```

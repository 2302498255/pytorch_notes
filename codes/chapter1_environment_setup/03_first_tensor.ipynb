## 1. 什么是张量？

张量（Tensor）是 PyTorch 中最基本的数据结构，可以理解为多维数组：

- **标量（0维张量）**: 单个数字，如 `5`
- **向量（1维张量）**: 数字序列，如 `[1, 2, 3]`
- **矩阵（2维张量）**: 数字表格，如 `[[1, 2], [3, 4]]`
- **高维张量**: 3维及以上，如图像批次、视频等

张量是深度学习的基础，所有的数据和模型参数都以张量形式存储。

```python
import torch
import numpy as np

print(f"PyTorch 版本: {torch.__version__}")
```

## 2. 创建张量的基本方法

### 2.1 从 Python 列表创建

```python
# 创建标量（0维张量）
scalar = torch.tensor(42)
print(f"标量: {scalar}")
print(f"形状: {scalar.shape}")
print(f"维度: {scalar.ndim}")
print()

# 创建向量（1维张量）
vector = torch.tensor([1, 2, 3, 4, 5])
print(f"向量: {vector}")
print(f"形状: {vector.shape}")
print(f"维度: {vector.ndim}")
print()

# 创建矩阵（2维张量）
matrix = torch.tensor([[1, 2, 3],
                       [4, 5, 6]])
print(f"矩阵:\n{matrix}")
print(f"形状: {matrix.shape}")
print(f"维度: {matrix.ndim}")
print()

# 创建3维张量（如RGB图像）
tensor_3d = torch.tensor([[[1, 2], [3, 4]],
                          [[5, 6], [7, 8]]])
print(f"3维张量:\n{tensor_3d}")
print(f"形状: {tensor_3d.shape}")
print(f"维度: {tensor_3d.ndim}")
```

### 2.2 使用特殊函数创建

```python
# 创建全零张量
zeros = torch.zeros(3, 4)
print(f"全零张量 (3x4):\n{zeros}")
print()

# 创建全一张量
ones = torch.ones(2, 3)
print(f"全一张量 (2x3):\n{ones}")
print()

# 创建单位矩阵
eye = torch.eye(4)
print(f"单位矩阵 (4x4):\n{eye}")
print()

# 创建随机张量（0到1之间的均匀分布）
rand = torch.rand(2, 3)
print(f"随机张量 (均匀分布):\n{rand}")
print()

# 创建随机张量（标准正态分布）
randn = torch.randn(2, 3)
print(f"随机张量 (正态分布):\n{randn}")
print()

# 创建等差数列
arange = torch.arange(0, 10, 2)  # 从0到10，步长为2
print(f"等差数列: {arange}")
print()

# 创建线性间隔张量
linspace = torch.linspace(0, 1, 5)  # 0到1之间均匀分布的5个数
print(f"线性间隔: {linspace}")
```

### 2.3 从 NumPy 数组创建

```python
# NumPy 数组转 PyTorch 张量
numpy_array = np.array([[1, 2, 3], [4, 5, 6]])
tensor_from_numpy = torch.from_numpy(numpy_array)

print(f"NumPy 数组:\n{numpy_array}")
print(f"类型: {type(numpy_array)}")
print()

print(f"PyTorch 张量:\n{tensor_from_numpy}")
print(f"类型: {type(tensor_from_numpy)}")
print()

# PyTorch 张量转 NumPy 数组
tensor = torch.tensor([[7, 8, 9], [10, 11, 12]])
numpy_from_tensor = tensor.numpy()

print(f"转回 NumPy:\n{numpy_from_tensor}")
print(f"类型: {type(numpy_from_tensor)}")

# 注意：共享内存
print("\n⚠️ 注意: from_numpy 创建的张量与原 NumPy 数组共享内存")
numpy_array[0, 0] = 999
print(f"修改 NumPy 数组后的张量:\n{tensor_from_numpy}")
```

## 5. 在不同设备上创建张量

这是本节的重点：学习如何在 CPU、CUDA、MPS 设备上创建和操作张量。

```python
# 获取可用设备
def get_available_device():
    """返回最优设备"""
    if torch.cuda.is_available():
        return torch.device('cuda')
    elif torch.backends.mps.is_available():
        return torch.device('mps')
    else:
        return torch.device('cpu')

device = get_available_device()

print("=" * 50)
print("在不同设备上创建张量")
print("=" * 50)

# 方法1: 创建时指定设备
print("\n方法1: 创建时指定设备")
tensor_cpu = torch.randn(3, 3, device='cpu')
print(f"CPU 张量: {tensor_cpu.device}")

if torch.cuda.is_available():
    tensor_cuda = torch.randn(3, 3, device='cuda')
    print(f"CUDA 张量: {tensor_cuda.device}")

if torch.backends.mps.is_available():
    tensor_mps = torch.randn(3, 3, device='mps')
    print(f"MPS 张量: {tensor_mps.device}")

# 方法2: 使用 .to() 方法迁移
print("\n方法2: 使用 .to() 方法迁移")
tensor = torch.randn(3, 3)  # 默认在 CPU
print(f"原始设备: {tensor.device}")

tensor_on_device = tensor.to(device)
print(f"迁移后设备: {tensor_on_device.device}")

# 方法3: 使用 .cuda() 或 .cpu() 快捷方法
print("\n方法3: 使用快捷方法")
tensor = torch.randn(3, 3)

if torch.cuda.is_available():
    tensor_cuda = tensor.cuda()  # 移到 CUDA
    print(f"使用 .cuda(): {tensor_cuda.device}")
    
    tensor_back = tensor_cuda.cpu()  # 移回 CPU
    print(f"使用 .cpu(): {tensor_back.device}")
```

## 6. 设备间数据迁移

在深度学习中，经常需要在不同设备间传输数据。

```python
print("=" * 50)
print("设备间数据迁移")
print("=" * 50)

# 创建 CPU 张量
cpu_tensor = torch.randn(1000, 1000)
print(f"\n原始张量设备: {cpu_tensor.device}")
print(f"张量形状: {cpu_tensor.shape}")

# 迁移到最优设备
device = get_available_device()
device_tensor = cpu_tensor.to(device)
print(f"\n迁移到: {device_tensor.device}")

# 重要：不同设备的张量不能直接运算
print("\n⚠️ 重要: 不同设备的张量不能直接运算")
try:
    # 这会报错（如果设备不同）
    result = cpu_tensor + device_tensor
except RuntimeError as e:
    print(f"错误: {e}")
    print("解决方法: 确保两个张量在同一设备上")
    
    # 正确的做法
    cpu_tensor_2 = torch.randn(1000, 1000)
    device_tensor_2 = cpu_tensor_2.to(device)
    result = device_tensor + device_tensor_2  # 两个都在同一设备
    print(f"成功计算，结果设备: {result.device}")
```

## 7. 设备操作性能测试

比较在不同设备上创建张量和进行计算的性能。

```python
import time

def benchmark_creation(device, size=5000, iterations=10):
    """
    测试在指定设备上创建张量的速度
    """
    times = []
    
    for _ in range(iterations):
        start = time.time()
        tensor = torch.randn(size, size, device=device)
        
        # 同步确保操作完成
        if device.type == 'cuda':
            torch.cuda.synchronize()
        elif device.type == 'mps':
            torch.mps.synchronize()
        
        end = time.time()
        times.append(end - start)
    
    return sum(times) / len(times)

def benchmark_computation(device, size=5000, iterations=10):
    """
    测试在指定设备上计算的速度
    """
    # 预先创建张量
    a = torch.randn(size, size, device=device)
    b = torch.randn(size, size, device=device)
    
    # 预热
    for _ in range(3):
        _ = torch.matmul(a, b)
    
    if device.type == 'cuda':
        torch.cuda.synchronize()
    elif device.type == 'mps':
        torch.mps.synchronize()
    
    times = []
    for _ in range(iterations):
        start = time.time()
        c = torch.matmul(a, b)
        
        if device.type == 'cuda':
            torch.cuda.synchronize()
        elif device.type == 'mps':
            torch.mps.synchronize()
        
        end = time.time()
        times.append(end - start)
    
    return sum(times) / len(times)

print("=" * 60)
print("设备性能对比")
print("=" * 60)

devices_to_test = [torch.device('cpu')]

if torch.cuda.is_available():
    devices_to_test.append(torch.device('cuda'))

if torch.backends.mps.is_available():
    devices_to_test.append(torch.device('mps'))

# 测试张量创建速度
print("\n【张量创建速度】(5000x5000 矩阵)")
creation_times = {}
for device in devices_to_test:
    avg_time = benchmark_creation(device)
    creation_times[str(device)] = avg_time
    print(f"{device}: {avg_time:.4f} 秒")

# 测试计算速度
print("\n【矩阵乘法速度】(5000x5000 矩阵)")
computation_times = {}
for device in devices_to_test:
    avg_time = benchmark_computation(device)
    computation_times[str(device)] = avg_time
    print(f"{device}: {avg_time:.4f} 秒")
    
    # 计算相对于 CPU 的加速比
    if device.type != 'cpu' and 'cpu' in computation_times:
        speedup = computation_times['cpu'] / avg_time
        print(f"  → 相比 CPU 加速: {speedup:.2f}x")

print("\n" + "=" * 60)
```

## 8. 设备迁移的性能影响

了解数据在设备间传输的开销。

```python
print("=" * 60)
print("设备迁移性能测试")
print("=" * 60)

# 创建大型张量
size = 5000
cpu_tensor = torch.randn(size, size)

print(f"\n张量大小: {size}x{size}")
print(f"内存占用: {cpu_tensor.element_size() * cpu_tensor.numel() / (1024**2):.2f} MB")

# 测试 CPU -> GPU 传输
if torch.cuda.is_available():
    print("\n【CPU -> CUDA 传输】")
    start = time.time()
    cuda_tensor = cpu_tensor.to('cuda')
    torch.cuda.synchronize()
    transfer_time = time.time() - start
    print(f"传输时间: {transfer_time:.4f} 秒")
    
    # 测试 GPU -> CPU 传输
    print("\n【CUDA -> CPU 传输】")
    start = time.time()
    cpu_tensor_back = cuda_tensor.to('cpu')
    transfer_time = time.time() - start
    print(f"传输时间: {transfer_time:.4f} 秒")

if torch.backends.mps.is_available():
    print("\n【CPU -> MPS 传输】")
    start = time.time()
    mps_tensor = cpu_tensor.to('mps')
    torch.mps.synchronize()
    transfer_time = time.time() - start
    print(f"传输时间: {transfer_time:.4f} 秒")
    
    # 测试 MPS -> CPU 传输
    print("\n【MPS -> CPU 传输】")
    start = time.time()
    cpu_tensor_back = mps_tensor.to('cpu')
    transfer_time = time.time() - start
    print(f"传输时间: {transfer_time:.4f} 秒")

print("\n💡 优化建议:")
print("  1. 尽量减少设备间数据传输")
print("  2. 批量传输数据而不是频繁小批量传输")
print("  3. 在训练时保持模型和数据在同一设备")
print("  4. 只在必要时（如保存结果）才传回 CPU")
```

## 10. 完整示例：设备无关的代码

```python
class DeviceAwareModel:
    """
    一个设备无关的简单模型示例
    可以自动适配 CPU/CUDA/MPS
    """
    def __init__(self):
        # 自动选择最优设备
        self.device = self._get_device()
        print(f"模型初始化在设备: {self.device}")
        
        # 创建模型参数（在正确的设备上）
        self.weight = torch.randn(10, 10, device=self.device)
        self.bias = torch.randn(10, device=self.device)
    
    def _get_device(self):
        """选择最优设备"""
        if torch.cuda.is_available():
            return torch.device('cuda')
        elif torch.backends.mps.is_available():
            return torch.device('mps')
        else:
            return torch.device('cpu')
    
    def forward(self, x):
        """
        前向传播
        自动将输入移到正确的设备
        """
        # 确保输入在正确的设备上
        x = x.to(self.device)
        
        # 计算
        output = torch.matmul(x, self.weight) + self.bias
        return output
    
    def get_device_info(self):
        """获取设备信息"""
        info = f"设备类型: {self.device.type}"
        
        if self.device.type == 'cuda':
            info += f"\nGPU 名称: {torch.cuda.get_device_name(self.device)}"
            memory = torch.cuda.get_device_properties(self.device).total_memory / (1024**3)
            info += f"\n显存: {memory:.2f} GB"
        
        return info

# 使用示例
print("=" * 60)
print("设备无关代码示例")
print("=" * 60)

# 创建模型（自动选择最优设备）
model = DeviceAwareModel()
print(f"\n{model.get_device_info()}")

# 创建输入数据（可以在任何设备上）
input_data = torch.randn(5, 10)  # CPU 上创建
print(f"\n输入数据设备: {input_data.device}")

# 前向传播（自动处理设备迁移）
output = model.forward(input_data)
print(f"输出数据设备: {output.device}")
print(f"输出形状: {output.shape}")

print("\n✅ 代码可以在任何设备上运行，无需修改！")
```

## 11. 常见错误和解决方法

```python
print("=" * 60)
print("常见错误和解决方法")
print("=" * 60)

print("""
错误1: RuntimeError: Expected all tensors to be on the same device
原因: 尝试对不同设备上的张量进行运算
解决: 使用 .to(device) 确保所有张量在同一设备

错误2: RuntimeError: CUDA out of memory
原因: GPU 显存不足
解决:
  - 减小 batch size
  - 使用梯度累积
  - 使用 torch.cuda.empty_cache() 清理缓存
  - 使用混合精度训练（float16）

错误3: RuntimeError: MPS backend out of memory
原因: MPS 设备内存不足
解决:
  - 减小模型或数据大小
  - 关闭其他占用 GPU 的应用
  - 回退到 CPU

错误4: 在 GPU 张量上调用 .numpy()
原因: NumPy 只支持 CPU 张量
解决: 先移到 CPU → tensor.cpu().numpy()

错误5: 性能比 CPU 还慢
原因: 数据太小或频繁传输
解决:
  - 对小数据集使用 CPU
  - 批量传输数据
  - 减少设备切换次数
""")
```

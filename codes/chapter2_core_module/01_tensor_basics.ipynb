```python
import torch
import numpy as np
```

## 1. 什么是张量

张量（Tensor）是 PyTorch 中的核心数据结构，可以看作是多维数组：
- 0维张量：标量（scalar）
- 1维张量：向量（vector）
- 2维张量：矩阵（matrix）
- 3维及以上：高维张量

```python
# 不同维度的张量示例
scalar = torch.tensor(3.14)  # 0维
vector = torch.tensor([1, 2, 3])  # 1维
matrix = torch.tensor([[1, 2], [3, 4]])  # 2维
tensor_3d = torch.tensor([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])  # 3维

print("标量 (0维):", scalar, "形状:", scalar.shape)
print("\n向量 (1维):", vector, "形状:", vector.shape)
print("\n矩阵 (2维):\n", matrix, "形状:", matrix.shape)
print("\n3维张量:\n", tensor_3d, "形状:", tensor_3d.shape)
```

## 2. 张量的基本属性

张量有八个主要属性：
1. **data**: 张量的数据
2. **dtype**: 数据类型
3. **shape**: 形状
4. **device**: 所在设备
5. **grad**: 梯度
6. **grad_fn**: 创建该张量的函数
7. **requires_grad**: 是否需要梯度
8. **is_leaf**: 是否是叶子节点

```python
# 创建一个张量
t = torch.tensor([[1., 2., 3.], [4., 5., 6.]])

print("=" * 60)
print("张量的基本属性")
print("=" * 60)
print(f"data: {t.data}")
print(f"dtype: {t.dtype}")
print(f"shape: {t.shape}")
print(f"device: {t.device}")
print(f"\n张量本身:\n{t}")
```

```python
# 设备转换示例
if torch.backends.mps.is_available() if hasattr(torch.backends, 'mps') else False:
    t_mps = t.to('mps')
    print(f"转移到 MPS 后的 device: {t_mps.device}")
elif torch.cuda.is_available():
    t_cuda = t.to('cuda')
    print(f"转移到 CUDA 后的 device: {t_cuda.device}")
else:
    print(f"当前 device: {t.device} (仅 CPU 可用)")
```

## 3. 梯度相关属性

梯度相关的四个属性对自动微分很重要：
- **requires_grad**: 是否需要计算梯度
- **is_leaf**: 是否是叶子节点（用户创建的张量）
- **grad**: 梯度值（反向传播后）
- **grad_fn**: 创建该张量的函数（非叶子节点才有）

```python
# 创建需要梯度的叶子节点
w = torch.tensor([1., 2., 3.], requires_grad=True)
x = torch.tensor([4., 5., 6.], requires_grad=True)

print("=" * 60)
print("叶子节点属性")
print("=" * 60)
print(f"w.requires_grad: {w.requires_grad}")
print(f"w.is_leaf: {w.is_leaf}")
print(f"w.grad: {w.grad}")
print(f"w.grad_fn: {w.grad_fn}")

# 进行运算
y = w * x + 2
print("\n" + "=" * 60)
print("运算后的非叶子节点属性")
print("=" * 60)
print(f"y.requires_grad: {y.requires_grad}")
print(f"y.is_leaf: {y.is_leaf}")
print(f"y.grad_fn: {y.grad_fn}")
print(f"y.grad: {y.grad}")

# 反向传播
loss = y.sum()
loss.backward()

print("\n" + "=" * 60)
print("反向传播后的梯度")
print("=" * 60)
print(f"w.grad: {w.grad}")
print(f"x.grad: {x.grad}")
print(f"y.grad: {y.grad}  (非叶子节点，梯度被释放)")
```

## 4. 综合示例：查看所有属性

```python
def print_tensor_info(name, tensor):
    """打印张量的所有属性信息"""
    print(f"\n{'='*60}")
    print(f"{name} 的属性信息")
    print(f"{'='*60}")
    print(f"data: {tensor.data}")
    print(f"dtype: {tensor.dtype}")
    print(f"shape: {tensor.shape}")
    print(f"device: {tensor.device}")
    print(f"requires_grad: {tensor.requires_grad}")
    print(f"is_leaf: {tensor.is_leaf}")
    print(f"grad_fn: {tensor.grad_fn}")
    print(f"grad: {tensor.grad}")

# 创建示例
a = torch.tensor([1., 2.], requires_grad=True)
b = torch.tensor([3., 4.], requires_grad=True)
c = a * b
d = c.sum()

print_tensor_info("叶子节点 a", a)
print_tensor_info("叶子节点 b", b)
print_tensor_info("中间节点 c", c)
print_tensor_info("输出节点 d", d)

# 反向传播
d.backward()

print("\n" + "="*60)
print("反向传播后")
print("="*60)
print(f"a.grad: {a.grad}")
print(f"b.grad: {b.grad}")
```

## 5. 常用数据类型

PyTorch 支持多种数据类型：

```python
# 整数类型
t_int8 = torch.tensor([1, 2, 3], dtype=torch.int8)
t_int32 = torch.tensor([1, 2, 3], dtype=torch.int32)
t_int64 = torch.tensor([1, 2, 3], dtype=torch.int64)

# 浮点类型
t_float16 = torch.tensor([1., 2., 3.], dtype=torch.float16)
t_float32 = torch.tensor([1., 2., 3.], dtype=torch.float32)
t_float64 = torch.tensor([1., 2., 3.], dtype=torch.float64)

# 布尔类型
t_bool = torch.tensor([True, False, True], dtype=torch.bool)

print("整数类型:")
print(f"int8:  {t_int8}, dtype: {t_int8.dtype}")
print(f"int32: {t_int32}, dtype: {t_int32.dtype}")
print(f"int64: {t_int64}, dtype: {t_int64.dtype}")

print("\n浮点类型:")
print(f"float16: {t_float16}, dtype: {t_float16.dtype}")
print(f"float32: {t_float32}, dtype: {t_float32.dtype}")
print(f"float64: {t_float64}, dtype: {t_float64.dtype}")

print("\n布尔类型:")
print(f"bool: {t_bool}, dtype: {t_bool.dtype}")
```

## 6. 关键要点

1. **data**: 存储张量的实际数据
2. **dtype**: 数据类型影响内存占用和计算精度
3. **shape**: 形状决定了张量的维度和大小
4. **device**: 张量可以在 CPU、CUDA、MPS 等设备间转移
5. **requires_grad=True**: 告诉 PyTorch 需要计算梯度
6. **is_leaf**: 用户创建的张量是叶子节点，运算产生的是非叶子节点
7. **grad**: 反向传播后存储梯度值（仅叶子节点保留）
8. **grad_fn**: 记录创建张量的运算（用于反向传播）

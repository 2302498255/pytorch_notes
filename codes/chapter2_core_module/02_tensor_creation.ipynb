# 2.1.2 张量的创建

本 notebook 介绍张量的各种创建方法：直接创建、依据数值创建、依概率分布创建

```python
import torch
import numpy as np
```

## 设备检查

```python
print("=" * 50)
print("PyTorch 设备检查")
print("=" * 50)
print(f"PyTorch 版本: {torch.__version__}")
print(f"\nCPU 可用: True (始终可用)")
print(f"CUDA 可用: {torch.cuda.is_available()}")
print(f"MPS 可用: {torch.backends.mps.is_available() if hasattr(torch.backends, 'mps') else False}")
print(f"MPS 已构建: {torch.backends.mps.is_built() if hasattr(torch.backends, 'mps') else False}")

# 确定可用设备
if torch.backends.mps.is_available() if hasattr(torch.backends, 'mps') else False:
    device = torch.device("mps")
    print(f"\n推荐使用设备: {device} (Apple Silicon GPU)")
elif torch.cuda.is_available():
    device = torch.device("cuda")
    print(f"\n推荐使用设备: {device} (NVIDIA GPU)")
else:
    device = torch.device("cpu")
    print(f"\n推荐使用设备: {device} (CPU)")

print("\n" + "=" * 50)
```

## 1. 直接创建

### 1.1 torch.tensor - 从数据直接创建

```python
# 从列表创建
l = [[1., -1.], [1., -1.]]
t_from_list = torch.tensor(l)
print("从列表创建:")
print(t_from_list, t_from_list.dtype)

# 从 numpy 数组创建
arr = np.array([[1, 2, 3], [4, 5, 6]])
t_from_array = torch.tensor(arr)
print("\n从 numpy 数组创建:")
print(t_from_array, t_from_array.dtype)

# 指定数据类型
t_from_array_uint8 = torch.tensor(arr, dtype=torch.uint8)
print("\n指定 dtype=torch.uint8:")
print(t_from_array_uint8, t_from_array_uint8.dtype)
```

### 1.2 torch.from_numpy - 从 numpy 数组创建（共享内存）

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
t_from_numpy = torch.from_numpy(arr)

print("初始状态:")
print("numpy array:", arr)
print("tensor     :", t_from_numpy)

print("\n修改 numpy 数组:")
arr[0, 0] = 0
print("numpy array:", arr)
print("tensor     :", t_from_numpy, "(也被修改了，共享内存)")

print("\n修改 tensor:")
t_from_numpy[0, 0] = -1
print("numpy array:", arr, "(也被修改了)")
print("tensor     :", t_from_numpy)
```

### 1.3 torch.as_tensor 和 torch.tensor 的区别

```python
arr = np.array([1, 2, 3])

# torch.tensor 会复制数据
t1 = torch.tensor(arr)
arr[0] = 100
print("torch.tensor (复制数据):")
print("修改 numpy 后，tensor 不变:", t1)

# torch.as_tensor 会共享内存（如果可能）
arr = np.array([1, 2, 3])
t2 = torch.as_tensor(arr)
arr[0] = 100
print("\ntorch.as_tensor (共享内存):")
print("修改 numpy 后，tensor 也变:", t2)
```

## 2. 依据数值创建

### 2.1 全0、全1张量

```python
# torch.zeros - 创建全0张量
zeros = torch.zeros(3, 3)
print("torch.zeros(3, 3):")
print(zeros)

# torch.ones - 创建全1张量
ones = torch.ones(2, 4)
print("\ntorch.ones(2, 4):")
print(ones)

# 使用 out 参数
o_t = torch.tensor([1])
t = torch.zeros((3, 3), out=o_t)
print("\n使用 out 参数（复用内存）:")
print(t)
print(f"t 和 o_t 是同一个对象: {id(t) == id(o_t)}")
```

### 2.2 根据已有张量创建

```python
t1 = torch.tensor([[1., -1.], [1., -1.]])
print("原始张量 t1:")
print(t1)

# zeros_like
t2 = torch.zeros_like(t1)
print("\ntorch.zeros_like(t1):")
print(t2)

# ones_like
t3 = torch.ones_like(t1)
print("\ntorch.ones_like(t1):")
print(t3)

# full_like
t4 = torch.full_like(t1, 3.14)
print("\ntorch.full_like(t1, 3.14):")
print(t4)
```

### 2.3 序列张量

```python
# torch.arange - 类似 Python range
t1 = torch.arange(0, 10, 2)
print("torch.arange(0, 10, 2):")
print(t1)

# torch.linspace - 线性等分
t2 = torch.linspace(3, 10, steps=5)
print("\ntorch.linspace(3, 10, steps=5):")
print(t2)

# torch.logspace - 对数等分
t3 = torch.logspace(start=0.1, end=1.0, steps=5)
print("\ntorch.logspace(start=0.1, end=1.0, steps=5):")
print(t3)

t4 = torch.logspace(start=2, end=2, steps=1, base=2)
print("\ntorch.logspace(start=2, end=2, steps=1, base=2):")
print(t4, "(2^2 = 4)")
```

### 2.4 特殊张量

```python
# torch.eye - 单位矩阵
eye = torch.eye(3)
print("torch.eye(3):")
print(eye)

# torch.full - 填充特定值
full = torch.full((2, 3), 7.)
print("\ntorch.full((2, 3), 7.):")
print(full)
```

## 3. 依概率分布创建

### 3.1 均匀分布

```python
# torch.rand - [0, 1) 均匀分布
rand = torch.rand(2, 3)
print("torch.rand(2, 3) - [0, 1) 均匀分布:")
print(rand)

# torch.randint - 整数均匀分布
randint = torch.randint(low=0, high=10, size=(3, 3))
print("\ntorch.randint(0, 10, (3, 3)) - [0, 10) 整数均匀分布:")
print(randint)

# torch.randperm - 随机排列
perm = torch.randperm(10)
print("\ntorch.randperm(10) - 0到9的随机排列:")
print(perm)
```

### 3.2 正态分布

```python
# torch.randn - 标准正态分布 N(0, 1)
randn = torch.randn(2, 3)
print("torch.randn(2, 3) - 标准正态分布 N(0, 1):")
print(randn)

# torch.normal - 指定均值和方差的正态分布
# 方式1: 所有元素使用相同的均值和标准差
normal = torch.normal(mean=0, std=1, size=(3, 4))
print("\ntorch.normal(mean=0, std=1, size=(3, 4)):")
print(normal)

# 方式2: 每个元素使用不同的均值和标准差
mean = torch.arange(1., 4.)
std = torch.arange(1., 4.)
normal2 = torch.normal(mean, std)
print("\ntorch.normal(mean=[1,2,3], std=[1,2,3]):")
print(normal2)
```

### 3.3 伯努利分布

```python
# torch.bernoulli - 伯努利分布（0或1）
prob = torch.tensor([0.5, 0.1, 1.])
bernoulli = torch.bernoulli(prob)
print("torch.bernoulli([0.5, 0.1, 1.]):")
print(bernoulli)

# 使用 uniform_ 生成概率矩阵
p = torch.empty(3, 3).uniform_(0, 1)
b = torch.bernoulli(p)
print("\n概率矩阵:")
print(p)
print("\n伯努利采样结果:")
print(b)
```

### 3.4 随机种子的重要性

```python
# 不设置随机种子，每次结果不同
print("不设置随机种子:")
print("第1次:", torch.rand(3))
print("第2次:", torch.rand(3))
print("第3次:", torch.rand(3))

print("\n设置随机种子:")
torch.manual_seed(42)
print("第1次:", torch.rand(3))
torch.manual_seed(42)
print("第2次:", torch.rand(3), "(相同)")
torch.manual_seed(42)
print("第3次:", torch.rand(3), "(相同)")
```

## 4. 总结

### 直接创建
- `torch.tensor()`: 从数据创建，复制数据
- `torch.from_numpy()`: 从 numpy 创建，共享内存
- `torch.as_tensor()`: 尽可能共享内存

### 依据数值创建
- `torch.zeros()`, `torch.ones()`, `torch.full()`: 特定值填充
- `torch.arange()`, `torch.linspace()`, `torch.logspace()`: 序列
- `torch.eye()`: 单位矩阵
- `*_like()`: 根据已有张量创建

### 依概率分布创建
- `torch.rand()`: [0,1) 均匀分布
- `torch.randn()`: 标准正态分布
- `torch.normal()`: 指定参数的正态分布
- `torch.randint()`: 整数均匀分布
- `torch.randperm()`: 随机排列
- `torch.bernoulli()`: 伯努利分布

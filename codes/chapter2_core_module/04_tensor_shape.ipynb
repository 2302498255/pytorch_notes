# 2.1.3 张量的形状变换

本 notebook 介绍张量的形状变换操作：reshape, view, permute, transpose, squeeze, unsqueeze

```python
import torch
import numpy as np
```

## 1. reshape - 改变形状

`reshape` 返回一个新形状的张量，可能会复制数据

```python
t = torch.arange(12)
print(f"原始张量: {t}")
print(f"形状: {t.shape}")

# reshape 成 3x4
t_reshaped = t.reshape(3, 4)
print(f"\nreshape(3, 4):\n{t_reshaped}")
print(f"形状: {t_reshaped.shape}")

# 使用 -1 自动推断
t_auto = t.reshape(-1, 4)
print(f"\nreshape(-1, 4) (自动推断第一维):\n{t_auto}")
print(f"形状: {t_auto.shape}")

t_auto2 = t.reshape(3, -1)
print(f"\nreshape(3, -1) (自动推断第二维):\n{t_auto2}")
print(f"形状: {t_auto2.shape}")
```

## 2. view - 视图变换

`view` 返回共享内存的新视图，要求张量在内存中是连续的

```python
t = torch.arange(12)
print(f"原始张量: {t}")

# view 成 3x4
t_view = t.view(3, 4)
print(f"\nview(3, 4):\n{t_view}")

# view 和 reshape 的区别：view 共享内存
print("\n测试共享内存:")
t[0] = 999
print(f"修改原始张量后，view 也变了:\n{t_view}")

# view 要求连续内存
print("\n\nview 要求连续内存:")
t2 = torch.arange(12).reshape(3, 4)
print(f"原始张量:\n{t2}")
print(f"is_contiguous: {t2.is_contiguous()}")

t2_t = t2.t()  # 转置
print(f"\n转置后:\n{t2_t}")
print(f"is_contiguous: {t2_t.is_contiguous()}")

try:
    t2_t.view(-1)  # 会失败
except RuntimeError as e:
    print(f"\nview 失败: {e}")

# 使用 contiguous() 解决
t2_t_cont = t2_t.contiguous()
print(f"\n使用 contiguous() 后可以 view:")
print(f"is_contiguous: {t2_t_cont.is_contiguous()}")
print(f"view(-1): {t2_t_cont.view(-1)}")
```

## 3. reshape vs view

- `view`: 必须是连续内存，返回视图（共享内存）
- `reshape`: 可以处理非连续内存，可能会复制数据

```python
t = torch.arange(12).reshape(3, 4)
print(f"原始张量:\n{t}")

t_t = t.t()  # 转置（非连续）
print(f"\n转置后 (非连续):\n{t_t}")
print(f"is_contiguous: {t_t.is_contiguous()}")

# reshape 可以处理非连续内存
t_reshape = t_t.reshape(-1)
print(f"\nreshape(-1) 成功: {t_reshape}")

# view 会失败
try:
    t_view = t_t.view(-1)
except RuntimeError as e:
    print(f"\nview(-1) 失败: view size is not compatible with input tensor's size and stride")
```

## 4. squeeze 和 unsqueeze - 压缩和扩展维度

- `squeeze`: 移除大小为 1 的维度
- `unsqueeze`: 在指定位置增加大小为 1 的维度

```python
# squeeze: 移除大小为 1 的维度
t = torch.randn(1, 3, 1, 4)
print(f"原始形状: {t.shape}")

# squeeze() 移除所有大小为 1 的维度
t_squeezed = t.squeeze()
print(f"squeeze() 后形状: {t_squeezed.shape}")

# squeeze(dim) 移除指定维度（如果大小为 1）
t_squeezed_dim0 = t.squeeze(dim=0)
print(f"squeeze(dim=0) 后形状: {t_squeezed_dim0.shape}")

t_squeezed_dim2 = t.squeeze(dim=2)
print(f"squeeze(dim=2) 后形状: {t_squeezed_dim2.shape}")

# 如果指定维度大小不是 1，不会改变
t_squeezed_dim1 = t.squeeze(dim=1)
print(f"squeeze(dim=1) 后形状: {t_squeezed_dim1.shape} (dim=1 大小为 3，不变)")
```

```python
# unsqueeze: 在指定位置增加维度
t = torch.randn(3, 4)
print(f"原始形状: {t.shape}")

# unsqueeze(dim=0) 在最前面增加维度
t_unsqueezed_0 = t.unsqueeze(dim=0)
print(f"unsqueeze(dim=0) 后形状: {t_unsqueezed_0.shape}")

# unsqueeze(dim=1) 在中间增加维度
t_unsqueezed_1 = t.unsqueeze(dim=1)
print(f"unsqueeze(dim=1) 后形状: {t_unsqueezed_1.shape}")

# unsqueeze(dim=2) 在最后增加维度
t_unsqueezed_2 = t.unsqueeze(dim=2)
print(f"unsqueeze(dim=2) 后形状: {t_unsqueezed_2.shape}")

# 使用负数索引
t_unsqueezed_neg1 = t.unsqueeze(dim=-1)
print(f"unsqueeze(dim=-1) 后形状: {t_unsqueezed_neg1.shape}")
```

## 5. transpose 和 permute - 维度转置

- `transpose`: 交换两个维度
- `permute`: 重新排列所有维度

```python
# transpose: 交换两个维度
t = torch.randn(2, 3, 4)
print(f"原始形状: {t.shape}")

# 交换 dim 0 和 dim 1
t_t01 = t.transpose(0, 1)
print(f"transpose(0, 1) 后形状: {t_t01.shape}")

# 交换 dim 1 和 dim 2
t_t12 = t.transpose(1, 2)
print(f"transpose(1, 2) 后形状: {t_t12.shape}")

# t() 是 transpose(0, 1) 的简写（仅适用于 2D 张量）
t_2d = torch.randn(3, 4)
print(f"\n2D 张量原始形状: {t_2d.shape}")
print(f"t() 后形状: {t_2d.t().shape}")
```

```python
# permute: 重新排列所有维度
t = torch.randn(2, 3, 4)
print(f"原始形状: {t.shape} (维度顺序: 0, 1, 2)")

# permute(2, 0, 1): dim2 -> dim0, dim0 -> dim1, dim1 -> dim2
t_permuted = t.permute(2, 0, 1)
print(f"\npermute(2, 0, 1) 后形状: {t_permuted.shape}")
print("解释: 原来的 dim2(4) -> 新的 dim0")
print("      原来的 dim0(2) -> 新的 dim1")
print("      原来的 dim1(3) -> 新的 dim2")

# permute(1, 2, 0)
t_permuted2 = t.permute(1, 2, 0)
print(f"\npermute(1, 2, 0) 后形状: {t_permuted2.shape}")
print("解释: 原来的 dim1(3) -> 新的 dim0")
print("      原来的 dim2(4) -> 新的 dim1")
print("      原来的 dim0(2) -> 新的 dim2")
```

## 6. flatten - 展平张量

```python
t = torch.randn(2, 3, 4)
print(f"原始形状: {t.shape}")

# flatten() 默认展平所有维度
t_flat = t.flatten()
print(f"\nflatten() 后形状: {t_flat.shape}")

# flatten(start_dim, end_dim) 展平指定范围的维度
t_flat_12 = t.flatten(start_dim=1)
print(f"flatten(start_dim=1) 后形状: {t_flat_12.shape}")
print("解释: 保留 dim0(2)，展平 dim1 和 dim2 (3*4=12)")

t_flat_01 = t.flatten(start_dim=0, end_dim=1)
print(f"\nflatten(start_dim=0, end_dim=1) 后形状: {t_flat_01.shape}")
print("解释: 展平 dim0 和 dim1 (2*3=6)，保留 dim2(4)")
```

## 7. 实际应用示例

### 7.1 批处理图像的形状变换

```python
# 假设图像批次: [batch, channels, height, width]
images = torch.randn(32, 3, 224, 224)
print(f"图像批次形状: {images.shape}")
print("[batch_size=32, channels=3, height=224, width=224]")

# 展平成全连接层的输入: [batch, features]
features = images.flatten(start_dim=1)
print(f"\n展平后形状: {features.shape}")
print(f"[batch_size=32, features={3*224*224}]")

# 转置通道: NCHW -> NHWC (PyTorch -> TensorFlow 格式)
images_nhwc = images.permute(0, 2, 3, 1)
print(f"\n转置后形状: {images_nhwc.shape}")
print("[batch_size=32, height=224, width=224, channels=3]")
```

### 7.2 添加批次维度

```python
# 单张图像: [channels, height, width]
single_image = torch.randn(3, 224, 224)
print(f"单张图像形状: {single_image.shape}")

# 添加批次维度: [1, channels, height, width]
batched_image = single_image.unsqueeze(0)
print(f"添加批次维度后: {batched_image.shape}")

# 或者使用 unsqueeze(dim=0)
batched_image2 = single_image[None, :, :, :]
print(f"使用切片添加批次维度: {batched_image2.shape}")
```

## 8. 总结

### 改变形状
- `reshape()`: 改变形状，可能复制数据
- `view()`: 改变形状，共享内存，要求连续
- `flatten()`: 展平张量

### 增减维度
- `squeeze()`: 移除大小为 1 的维度
- `unsqueeze()`: 增加大小为 1 的维度

### 转置维度
- `transpose()`: 交换两个维度
- `permute()`: 重新排列所有维度
- `t()`: 2D 张量转置的简写

### 关键要点
1. `view` 要求连续内存，`reshape` 更灵活
2. `transpose` 和 `permute` 返回的张量是非连续的
3. 使用 `contiguous()` 可以将非连续张量变为连续
4. `unsqueeze` 常用于添加批次维度
5. `flatten` 常用于将特征图展平成向量

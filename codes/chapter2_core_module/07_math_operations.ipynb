# 2.1.6 张量的数学运算

本 notebook 介绍张量的各种数学运算：逐元素操作、聚合操作、比较操作、线性代数等

```python
import torch
import numpy as np
```

## 1. 逐元素操作（Element-wise Operations）

```python
# 逐元素操作：对每个元素独立进行运算
x = torch.tensor([-1.0, 0.5, 2.0])

print("原始张量:")
print(f"x: {x}")
print("\n逐元素操作:")
print(f"abs(x)     : {torch.abs(x)}")
print(f"relu(x)    : {torch.relu(x)}")
print(f"clamp(x,0,1): {torch.clamp(x, 0., 1.)}")
print(f"pow(x, 2)  : {torch.pow(x, 2)}")
print(f"sqrt(|x|)  : {torch.sqrt(torch.abs(x))}")
print(f"exp(x)     : {torch.exp(x)}")
print(f"log(|x|+1) : {torch.log(torch.abs(x) + 1)}")
```

```python
# 三角函数
x = torch.tensor([0., np.pi/4, np.pi/2, np.pi])
print("三角函数:")
print(f"x        : {x}")
print(f"sin(x)   : {torch.sin(x)}")
print(f"cos(x)   : {torch.cos(x)}")
print(f"tan(x)   : {torch.tan(x)}")
```

## 2. 算术运算

```python
# 基本算术运算
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])

print("基本算术运算:")
print(f"a: {a}")
print(f"b: {b}")
print(f"\na + b   : {a + b}")
print(f"a - b   : {a - b}")
print(f"a * b   : {a * b}  (逐元素乘法)")
print(f"a / b   : {a / b}")
print(f"a // b  : {a // b}  (整除)")
print(f"a % 2   : {a % 2}  (取余)")
print(f"a ** 2  : {a ** 2}  (幂运算)")
```

## 3. 重要区分：mul vs mm vs matmul

- `mul`: 逐元素乘法（Element-wise multiplication）
- `mm`: 矩阵乘法（Matrix multiplication）
- `matmul`: 通用矩阵乘法（支持广播）

```python
# mul: 逐元素乘法
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6], [7, 8]])

print("mul - 逐元素乘法:")
print(f"a:\n{a}")
print(f"\nb:\n{b}")
print(f"\na * b (或 torch.mul(a, b)):\n{a * b}")
print("解释: 对应位置元素相乘")

# mm: 矩阵乘法
print("\n" + "="*60)
print("mm - 矩阵乘法:")
result_mm = torch.mm(a, b)
print(f"torch.mm(a, b):\n{result_mm}")
print("解释: 标准矩阵乘法，a的列数必须等于b的行数")
```

```python
# matmul: 通用矩阵乘法
print("matmul - 通用矩阵乘法:")
print("="*60)

# 2D x 2D: 等同于 mm
a = torch.randn(2, 3)
b = torch.randn(3, 4)
result = torch.matmul(a, b)
print(f"2D x 2D: {a.shape} x {b.shape} -> {result.shape}")

# 1D x 1D: 点积
a = torch.randn(3)
b = torch.randn(3)
result = torch.matmul(a, b)
print(f"\n1D x 1D: {a.shape} x {b.shape} -> {result.shape} (标量)")

# 批量矩阵乘法
a = torch.randn(10, 2, 3)
b = torch.randn(10, 3, 4)
result = torch.matmul(a, b)
print(f"\n批量: {a.shape} x {b.shape} -> {result.shape}")

# 广播
a = torch.randn(10, 2, 3)
b = torch.randn(3, 4)
result = torch.matmul(a, b)
print(f"\n广播: {a.shape} x {b.shape} -> {result.shape}")
```

## 4. 聚合操作（Reduction Operations）

```python
# 聚合操作：将多个值聚合成一个或少数几个值
x = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]])

print(f"原始张量:\n{x}")
print(f"形状: {x.shape}")

print("\n聚合操作:")
print(f"sum (全部)     : {x.sum()}")
print(f"sum (dim=0)    : {x.sum(dim=0)}  (列求和)")
print(f"sum (dim=1)    : {x.sum(dim=1)}  (行求和)")

print(f"\nmean (全部)    : {x.mean()}")
print(f"mean (dim=0)   : {x.mean(dim=0)}")
print(f"mean (dim=1)   : {x.mean(dim=1)}")

print(f"\nmax (全部)     : {x.max()}")
print(f"max (dim=0)    : {x.max(dim=0)}")
print(f"max (dim=1)    : {x.max(dim=1)}")

print(f"\nmin (全部)     : {x.min()}")
print(f"argmax (全部)  : {x.argmax()}")
print(f"argmax (dim=1) : {x.argmax(dim=1)}")
```

```python
# keepdim: 保持维度
x = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]])

print(f"原始形状: {x.shape}")
print(f"\nsum(dim=1) 形状: {x.sum(dim=1).shape}")
print(f"sum(dim=1, keepdim=True) 形状: {x.sum(dim=1, keepdim=True).shape}")

print("\nkeepdim=True 的作用: 便于广播")
sum_keepdim = x.sum(dim=1, keepdim=True)
print(f"x / sum(dim=1, keepdim=True):\n{x / sum_keepdim}")
```

## 5. 比较操作

```python
# 比较操作
x = torch.tensor([[1.0, -2.0, 3.0],
                  [0.5,  4.0, -1.0]])

print(f"x:\n{x}")

# 逐元素比较
print("\n逐元素比较:")
print(f"x > 0:\n{x > 0}")
print(f"x == 0:\n{x == 0}")
print(f"x < 0:\n{x < 0}")

# where: 条件选择
mask = x > 0
result = torch.where(mask, x, torch.zeros_like(x))
print(f"\nwhere(x > 0, x, 0):\n{result}")
print("解释: 大于0保留，否则置0")
```

```python
# topk: 找出最大的 k 个值
x = torch.tensor([[1.0, -2.0, 3.0, 5.0, -1.0],
                  [0.5,  4.0, -1.0, 2.0, 3.5]])

values, indices = torch.topk(x, k=3, dim=1)
print(f"x:\n{x}")
print(f"\ntop-3 值:\n{values}")
print(f"\ntop-3 索引:\n{indices}")

# sort: 排序
sorted_values, sorted_indices = torch.sort(x, dim=1, descending=True)
print(f"\n排序后的值:\n{sorted_values}")
print(f"\n排序后的索引:\n{sorted_indices}")
```

## 6. 线性代数运算

```python
# 向量点积
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])
dot_product = torch.dot(a, b)
print("向量点积:")
print(f"a: {a}")
print(f"b: {b}")
print(f"dot(a, b): {dot_product}")
print(f"验证: {1*4 + 2*5 + 3*6} = {dot_product}")

# 矩阵-向量乘法
print("\n" + "="*60)
print("矩阵-向量乘法:")
A = torch.randn(3, 4)
x = torch.randn(4)
result = torch.mv(A, x)
print(f"A 形状: {A.shape}")
print(f"x 形状: {x.shape}")
print(f"mv(A, x) 形状: {result.shape}")

# 矩阵-矩阵乘法
print("\n" + "="*60)
print("矩阵-矩阵乘法:")
A = torch.randn(2, 3)
B = torch.randn(3, 4)
result = torch.mm(A, B)
print(f"A 形状: {A.shape}")
print(f"B 形状: {B.shape}")
print(f"mm(A, B) 形状: {result.shape}")

# 批量矩阵乘法
print("\n" + "="*60)
print("批量矩阵乘法:")
A = torch.randn(10, 2, 3)
B = torch.randn(10, 3, 4)
result = torch.bmm(A, B)
print(f"A 形状: {A.shape}")
print(f"B 形状: {B.shape}")
print(f"bmm(A, B) 形状: {result.shape}")
```

```python
# 范数
x = torch.tensor([[1., 2., 3.],
                  [4., 5., 6.]])

print(f"x:\n{x}")
print("\n范数计算:")
print(f"L2 范数 (全部): {torch.norm(x)}")
print(f"L2 范数 (dim=1): {torch.norm(x, dim=1)}")
print(f"L1 范数 (全部): {torch.norm(x, p=1)}")
print(f"无穷范数 (全部): {torch.norm(x, p=float('inf'))}")
```

## 7. 实际应用示例

### 7.1 Softmax 实现

```python
# 手动实现 softmax
logits = torch.randn(3, 5)
print(f"logits:\n{logits}")

# 方法1: 使用 torch.softmax
softmax_builtin = torch.softmax(logits, dim=1)
print(f"\ntorch.softmax(dim=1):\n{softmax_builtin}")

# 方法2: 手动实现
exp_logits = torch.exp(logits)
softmax_manual = exp_logits / exp_logits.sum(dim=1, keepdim=True)
print(f"\n手动实现:\n{softmax_manual}")

# 验证
print(f"\n结果相同: {torch.allclose(softmax_builtin, softmax_manual)}")
print(f"每行的和: {softmax_manual.sum(dim=1)}")
```

### 7.2 L2 正则化

```python
# L2 正则化示例
weights = torch.randn(10, 5, requires_grad=True)

# 计算 L2 正则化项
l2_reg = torch.norm(weights, p=2) ** 2
print(f"权重形状: {weights.shape}")
print(f"L2 正则化项: {l2_reg}")

# 或者
l2_reg2 = (weights ** 2).sum()
print(f"另一种计算方式: {l2_reg2}")
print(f"结果相同: {torch.allclose(l2_reg, l2_reg2)}")
```

### 7.3 余弦相似度

```python
# 余弦相似度
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])

# 方法1: 使用 torch.nn.functional
import torch.nn.functional as F
cos_sim_builtin = F.cosine_similarity(a.unsqueeze(0), b.unsqueeze(0))
print(f"a: {a}")
print(f"b: {b}")
print(f"\ntorch.nn.functional.cosine_similarity: {cos_sim_builtin}")

# 方法2: 手动实现
cos_sim_manual = torch.dot(a, b) / (torch.norm(a) * torch.norm(b))
print(f"手动实现: {cos_sim_manual}")

print(f"\n结果相同: {torch.allclose(cos_sim_builtin, cos_sim_manual)}")
```

## 8. 总结

### 逐元素操作
- `abs`, `sqrt`, `exp`, `log`, `sin`, `cos`
- `relu`, `clamp`, `pow`

### 算术运算
- `+`, `-`, `*`, `/`, `**`
- **重要**: `*` 是逐元素乘法，矩阵乘法用 `mm` 或 `matmul`

### 聚合操作
- `sum`, `mean`, `max`, `min`
- `argmax`, `argmin`
- 使用 `keepdim=True` 保持维度

### 比较操作
- `>`, `<`, `==`, `>=`, `<=`
- `where`, `topk`, `sort`

### 线性代数
- `dot`: 向量点积
- `mv`: 矩阵-向量乘法
- `mm`: 矩阵-矩阵乘法
- `bmm`: 批量矩阵乘法
- `matmul`: 通用矩阵乘法（支持广播）
- `norm`: 范数计算

### 关键要点
1. 区分 `mul`（逐元素）和 `mm`（矩阵乘法）
2. `dim` 参数控制在哪个维度上聚合
3. `keepdim=True` 便于后续广播
4. 大多数操作都有 in-place 版本（加 `_` 后缀）
